```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go 
```

```python
file = "../final_sdf.csv"
```

```python
df = pd.read_csv(file)
df.head()
```

```python
fig = go.Figure(data=go.Isosurface(
    x=df.x,
    y=df.y,
    z=df.z,
    value=df.value,
    surface_count=4,
    slices= {
        'x': {'show': True, 'locations': [df.x.max()]},
        'y': {'show': True, 'locations': [df.y.min()/2]},
        'z': {'show': True, 'locations': [df.z.mean()]}
    },
    caps=dict(x_show=False, y_show=False, z_show=False),
    colorscale='Viridis',
    opacity=0.6
))

# Add axis labels and title
fig.update_layout(
    title='3D Volumetric Slice Visualization',
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    )
)
fig.show()
```
